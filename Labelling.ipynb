```python
import torch
import cv2
import numpy as np
import os

root_dir = "clothes"

# Load the YOLOv5 object detection model
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')

# Define a function to label an image based on the presence of clothes
def label_image(image):
    # Apply the YOLOv5 model to the image
    results = model(image)
    
    # Extract the bounding boxes and labels from the results
    boxes = results.xyxy[0]
    labels = results.xyxyn[0][:, -1]

    # Check if any of the labels correspond to pants, shirts or tshirts
    labels_dict = {2: "pants", 3: "shirt", 4: "tshirt"}
    clothes_labels = [labels_dict[int(label)] for label in labels if int(label) in labels_dict]
    if len(clothes_labels) == 0:
        return None
    else:
        return clothes_labels[0]

# Traverse the directory tree and label the images
for subdir in ["pants", "shirts", "tshirt"]:
    # Construct the path to the subdirectory
    sub_dir = os.path.join(root_dir, subdir)

    # Get the list of image files in the subdirectory
    image_files = [os.path.join(sub_dir, f) for f in os.listdir(sub_dir) if f.endswith(".jpg")]

    # Label each image
    for image_path in image_files:
        # Load the image
        image = cv2.imread(image_path)

        # Convert the image to RGB format
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        # Label the image
        label = label_image(image)

        # Save the label to a file
        if label is not None:
            label_path = image_path.replace(".jpg", ".txt")
            with open(label_path, "w") as f:
                f.write(label)
```

```python
import torch
import cv2
import numpy as np
import os

root_dir = "clothes"

# Load the YOLOv5 object detection model
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')

# Define a function to label an image based on the presence of clothes
def label_image(image):
    # Apply the YOLOv5 model to the image
    results = model(image)
    
    # Extract the bounding boxes and labels from the results
    boxes = results.xyxy[0]
    labels = results.xyxyn[0][:, -1]

    # Check if any of the labels correspond to pants, shirts or tshirts
    labels_dict = {2: "pants", 3: "shirt", 4: "tshirt"}
    clothes_labels = [labels_dict[int(label)] for label in labels if int(label) in labels_dict]
    if len(clothes_labels) == 0:
        return None
    else:
        return clothes_labels[0]

    
# Traverse the directory tree and label the images
for subdir in ["pants", "shirts", "tshirt"]:
    # Construct the path to the subdirectory
    sub_dir = os.path.join(root_dir, subdir)

    # Get the list of image files in the subdirectory
    image_files = [os.path.join(sub_dir, f) for f in os.listdir(sub_dir) if f.endswith(".jpg")]

    # Label each image
    for image_path in image_files:
        # Load the image
        image = cv2.imread(image_path)

        # Convert the image to RGB format
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        # Label the image
        label = label_image(image)

        # Save the label to a file
        label_path = image_path.replace(".jpg", ".txt")
        with open(label_path, "w") as f:
            f.write(subdir)
```
